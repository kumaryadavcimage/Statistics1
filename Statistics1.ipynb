#### Q2. What are the types of Probability distribution?

#### solve
There are several types of probability distributions, and they can be broadly categorized into two main classes: discrete probability distributions and continuous probability distributions.

Discrete Probability Distributions:

Bernoulli Distribution: Describes a random variable with only two possible outcomes, usually labeled as success and failure.

Binomial Distribution: Represents the number of successes in a fixed number of independent Bernoulli trials.

Poisson Distribution: Models the number of events that occur within a fixed interval of time or space, assuming a constant average rate.

Continuous Probability Distributions:

Normal Distribution (Gaussian Distribution): A symmetric, bell-shaped distribution that is characterized by its mean and standard deviation. Many natural phenomena follow a normal distribution.

Uniform Distribution: All outcomes are equally likely, and the probability density function is constant over the range of possible values.

Exponential Distribution: Describes the time between events in a Poisson process, where events occur continuously and independently at a constant average rate.

Gamma Distribution: Generalizes the exponential distribution to account for waiting times until a specified number of events occur.

Beta Distribution: Represents a family of continuous probability distributions defined on the interval [0, 1]. It is often used to model the distribution of random variables limited to this range.

Cauchy Distribution: Characterized by its heavy tails, and it does not have a defined mean or variance.

Log-Normal Distribution: If the natural logarithm of a random variable follows a normal distribution, then the variable itself follows a log-normal distribution.

#### Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.


#### solve
you can use the probability density function(PDF) formula for the normal distribution, which is given by:
    
f(x;mu,sigma) =1/sigma root 2pi . log -1/2(x-mu/sigma)^2


Here's a Python function that calculates the PDF of a normal distribution at a given point:

```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution at a given point.

    Parameters:
    - x: The value at which to calculate the PDF.
    - mean: The mean of the normal distribution.
    - std_dev: The standard deviation of the normal distribution.

    Returns:
    The PDF at the given point.
    """
    if std_dev <= 0:
        raise ValueError("Standard deviation must be positive.")

    exponent = -0.5 * ((x - mean) / std_dev)**2
    pdf = (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(exponent)

    return pdf

# Example usage:
mean = 0
std_dev = 1
point = 1.5

pdf_value = normal_pdf(point, mean, std_dev)
print(f"The PDF at x={point} is: {pdf_value}")
```

#### Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.


#### solve

The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes (success or failure). The properties of the binomial distribution include:

a.Number of Trials (n): The distribution is characterized by the number of trials (n), which represents the fixed number of independent experiments or trials.

b.Probability of Success (p): Each trial has a probability of success denoted by p, and the probability of failure is 1-p

Independence: The trials are assumed to be independent, meaning the outcome of one trial does not affect the outcome of another.

Binary Outcome: Each trial results in a binary outcome (success or failure).

Discreteness: The random variable representing the number of successes is discrete, taking on integer values.

Probability Mass Function (PMF): The probability mass function of the binomial distribution is given by:

p(x=k)=(n to k).p^k.(1-p)^n-k



#### Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.


#### solve
Certainly! You can use the numpy library to generate a random sample from a binomial distribution and matplotlib for plotting the histogram. Here's a Python script that demonstrates this:

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.4  # Probability of success

# Generate a random sample of size 1000 from the binomial distribution
sample = np.random.binomial(n, p, size=1000)

# Plotting the histogram
plt.hist(sample, bins=np.arange(-0.5, n + 1.5, 1), edgecolor='black', alpha=0.7)
plt.title(f"Binomial Distribution (n={n}, p={p}) - Random Sample of Size 1000")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.


#### solve

Certainly! The Cumulative Distribution Function (CDF) for a Poisson distribution with a given mean λ at a given point k is given by:
    
p(x<equal k) =sigma (i=0 to k) e-lambda.lambda^i/i!

Here's a Python function that calculates the CDF of a Poisson distribution:

```python
import math

def poisson_cdf(k, mean):
    """
    Calculate the Cumulative Distribution Function (CDF) of a Poisson distribution at a given point.

    Parameters:
    - k: The value at which to calculate the CDF.
    - mean: The mean of the Poisson distribution.

    Returns:
    The CDF at the given point.
    """
    if k < 0 or mean <= 0:
        raise ValueError("Both k and mean must be non-negative.")

    cdf = 0
    for i in range(k + 1):
        cdf += math.exp(-mean) * (mean**i) / math.factorial(i)

    return cdf

# Example usage:
mean = 3
point = 2

cdf_value = poisson_cdf(point, mean)
print(f"The CDF at k={point} is: {cdf_value}")
```

#### Q7. How Binomial distribution different from Poisson distribution?

#### solve
The Binomial distribution and the Poisson distribution are both discrete probability distributions, but they are used to model different types of random phenomena and have distinct characteristics.

a.Nature of the Events:

Binomial Distribution: Models the number of successes in a fixed number of independent Bernoulli trials, where each trial can result in only two outcomes (success or failure).

Poisson Distribution: Models the number of events that occur in a fixed interval of time or space, with events happening randomly and independently.

b.Number of Trials:

Binomial Distribution: Requires a fixed number of trials (n) for which the probability of success (p) is constant.

Poisson Distribution: Does not involve a fixed number of trials; it describes the number of events in a continuous or large number of trials.

c.Probability of Success:

Binomial Distribution: Each trial has a constant probability of success (p) and failure (1 − 1−p).

Poisson Distribution: Describes events that occur with a low probability (p) in each trial, but the overall rate of events is constant.

d.Conditions:

Binomial Distribution: Applicable when there is a fixed number of trials, and the trials are independent with a constant probability of success.

Poisson Distribution: Applicable when events are rare and random, and the number of trials is large or continuous.

#### Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.


#### solve

Certainly! You can use the numpy library to generate a random sample from a Poisson distribution and calculate the sample mean and variance. Here's a Python script that demonstrates this:

```python
import numpy as np

# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the Poisson distribution
mean = 5

# Generate a random sample of size 1000 from the Poisson distribution
sample = np.random.poisson(mean, size=1000)

# Calculate the sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample, ddof=1)  # using ddof=1 for sample variance

# Print the results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```
